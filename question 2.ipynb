```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Date reported'] = pd.to_datetime(df['Date reported'])
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Date reported', y='Valuation', hue='Company')
plt.title("Valuation Over Time")
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Amount', y='Valuation', hue='Industry')
plt.title("Amount vs. Valuation")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Amount', bins=20)
plt.title("Amount Distribution")
```

```python
industry_counts = df['Industry'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%')
plt.title("Industry Distribution")
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Lead investors', order=df['Lead investors'].value_counts().index[:10])
plt.title("Top Lead Investors")
```

```python
df['Month'] = df['Date reported'].dt.month
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Month')
plt.title("Investments per Month")
```

```python
lead_investors_counts = df['Lead investors'].str.count(',') + 1  # Count the number of lead investors
plt.figure(figsize=(12, 6))
sns.barplot(x=lead_investors_counts, y=df['Company'], ci=None)
plt.title("Number of Lead Investors")
plt.xlabel('Number of Lead Investors')
plt.ylabel('Company')
```

```python
biotech_df = df[df['Industry'] == 'Biotech']
plt.figure(figsize=(10, 5))
sns.lineplot(data=biotech_df, x='Date reported', y='Valuation', hue='Company', marker='o')
plt.title("Valuation Over Time in Biotech Industry")
```

```python
df['Date reported'] = pd.to_datetime(df['Date reported'], errors='coerce')
df['Year'] = df['Date reported'].dt.year
total_investment_by_year = df.groupby('Year')['Amount'].sum()
plt.figure(figsize=(10, 5))
sns.lineplot(data=total_investment_by_year)
plt.title("Total Investment Amount Over Time")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Lead investors', hue='Industry')
plt.xticks(rotation=45)
plt.title("Lead Investors by Industry")

```

```python
plt.figure(figsize=(10, 5))
sns.swarmplot(data=df, x='Industry', y='Amount')
plt.xticks(rotation=45)
plt.title("Amount Distribution by Industry (Swarm Plot)")
```

```python
space_df = df[df['Industry'] == 'Space']
plt.figure(figsize=(10, 5))
sns.lineplot(data=space_df, x='Date reported', y='Amount', hue='Company')
plt.title("Amount Over Time in Space Industry")
```
